## Vacuum Gauge Explorer
Interactive tool for stepping through different gauges across many fills. Features include:
* Single key-press navigation 
    * Step Forward / Backward
    * Jump to Specific Fill
    * Save favourite location(s)
    * Automatic step through X fills with pre-specified location
* On-the-fly use of loaded classifiers
    * Load models of choice 
    * Visualize analytical approaches
* Multi-plotting of gauges
* Automatic saving of loaded figures and data (quicker access in subsequent runs)
